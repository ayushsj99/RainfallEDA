```python
import pandas as pd
import seaborn as sns
import plotly.graph_objects as go
import plotly.express as px
import matplotlib.pyplot as plt
%matplotlib inline
sns.set_theme(style="darkgrid")
rainfalldf = pd.read_csv("rainfall_area-wt_india_1901-2015.csv")
rainfalldf.head()
```

```python
rainfalldf.info()
```

```python
# Using Seaborn
sns.set_context("notebook")  # Adjust context styling
annual=sns.relplot(
    x='YEAR', 
    y='ANNUAL', 
    data=rainfalldf, 
    kind='line', 
    height=8,  # Height of the plot
    aspect=1.5  # Aspect ratio (width = height * aspect)
)

annual.ax.axhline(rainfalldf['ANNUAL'].mean(), color='red', linestyle="--", linewidth=1.5, label="Mean Rainfall")
annual.ax.legend()
plt.show()
```
